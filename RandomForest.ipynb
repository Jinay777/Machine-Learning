# Part 1

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
X = iris.iloc[:, :-1]
y = iris.iloc[:, -1]
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X, y, test_size=0.3, random_state = 1)
X2_train, X2_test, y2_train, y2_test = train_test_split(X, y, test_size=0.3, random_state = 2)
X3_train, X3_test, y3_train, y3_test = train_test_split(X, y, test_size=0.3, random_state = 3)
```

```python
clf1 = DecisionTreeClassifier(max_depth =3, random_state = 42)
clf2 = DecisionTreeClassifier(max_depth =3, random_state = 42)
clf3 = DecisionTreeClassifier(max_depth =3, random_state = 42)
```

```python
clf1.fit(X1_train, y1_train)
y1_pred = clf1.predict(X1_test)
print(f'Accuracy score of DT 1 is {accuracy_score(y1_test, y1_pred)}')
clf1.fit(X1_train, y1_train)
y1_pred = clf1.predict(X1_test)
print(f'Accuracy score of DT 1 is {accuracy_score(y1_test, y1_pred)}')

clf2.fit(X2_train, y2_train)
y2_pred = clf2.predict(X2_test)
print(f'Accuracy score of DT 2 is {accuracy_score(y2_test, y2_pred)}')

clf3.fit(X3_train, y3_train)
y3_pred = clf3.predict(X3_test)
print(f'Accuracy score of DT 3 is {accuracy_score(y3_test, y3_pred)}')
clf2.fit(X2_train, y2_train)
y2_pred = clf2.predict(X2_test)
print(f'Accuracy score of DT 2 is {accuracy_score(y2_test, y2_pred)}')

clf3.fit(X3_train, y3_train)
y3_pred = clf3.predict(X3_test)
print(f'Accuracy score of DT 3 is {accuracy_score(y3_test, y3_pred)}')
```

```python
plot_tree(clf1, class_names = iris.columns, filled = True);
```

```python
plot_tree(clf2, class_names = iris.columns, filled = True);
```

```python
plot_tree(clf3, class_names = iris.columns, filled = True);
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(f'Accuracy score of RF is {accuracy_score(y_test, y_pred)}')
```

```python
importances = clf.feature_importances_

sorted_indices = np.argsort(importances)[::-1]

feat_labels = iris.columns[:-1]

selected_features = []

for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30,
                            feat_labels[sorted_indices[f]],
                            importances[sorted_indices[f]]))

for f in range(2):
  selected_features.append(feat_labels[sorted_indices[f]])

print(f'\n\n Selected features are {selected_features}')
```

```python
X_new = iris[selected_features]
y_new = iris['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, test_size=0.3, random_state = 42)
```

```python
clf1 = DecisionTreeClassifier(max_depth =3, random_state = 42)
clf2 = DecisionTreeClassifier(max_depth =3, random_state = 42)
clf3 = DecisionTreeClassifier(max_depth =3, random_state = 42)
```

```python
clf1.fit(X1_train, y1_train)
y1_pred = clf1.predict(X1_test)
print(f'Accuracy score of DT 1 is {accuracy_score(y1_test, y1_pred)}')

clf2.fit(X2_train, y2_train)
y2_pred = clf2.predict(X2_test)
print(f'Accuracy score of DT 2 is {accuracy_score(y2_test, y2_pred)}')

clf3.fit(X3_train, y3_train)
y3_pred = clf3.predict(X3_test)
print(f'Accuracy score of DT 3 is {accuracy_score(y3_test, y3_pred)}')
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(f'Accuracy score of RF is {accuracy_score(y_test, y_pred)}')
```
